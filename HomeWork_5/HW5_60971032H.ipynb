```python
import jieba
import re
import logging
import os #基本上這兩行非常非常重要，可以處裡奇怪的系統問題
import tempfile #基本上這兩行非常非常重要，可以調用TEMP裡面的暫存空間
import pyLDAvis.gensim
from gensim import corpora
```

```python
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
TEMP_FOLDER = tempfile.gettempdir()
print('文件夹"{}" 會被拿來存該死的字典'.format(TEMP_FOLDER))
#這幾行沒打，怎麼set都會發現corpus永遠是空的
```

```python
file_path = "text1.txt"
with open(file_path, "r", encoding="utf-8") as file:
    text1 = file.read()
# 移除換行、空行等雜亂資訊
text1 = re.sub(r"\s+", "", text1)
```

```python
file_path = "text2.txt"
with open(file_path, "r", encoding="utf-8") as file:
    text2 = file.read()
# 移除換行、空行等雜亂資訊
text2 = re.sub(r"\s+", "", text2)
```

```python
file_path = "text3.txt"
with open(file_path, "r", encoding="utf-8") as file:
    text3 = file.read()
# 移除換行、空行等雜亂資訊
text3 = re.sub(r"\s+", "", text3)
```

```python
file_path = "text4.txt"
with open(file_path, "r", encoding="utf-8") as file:
    text4 = file.read()
# 上面四行的寫法我可以用迴圈，但我累了...........
text4 = re.sub(r"\s+", "", text4)
```

```python
raw_corpus = [text1,text2,text3,text4] #處理之前的資訊，基本上就是沒設停用字的訊息
```

```python
corpus = []
for sentence in raw_corpus:
    sentence = ''.join(re.findall(r'[\u4e00-\u9fa5]+', sentence))  # 把英文等其他的字元砍掉
    stop_words = ['我','啊','他','個','是','們','的','她','說','會','不']  #設定停用字詞
    corpus.append([item for item in jieba.cut(sentence) if item not in stop_words])  # 去掉上面stop_words裡面的東西，之後輸出corpus
```

```python
dictionary = corpora.Dictionary(corpus)
dictionary.save(os.path.join(TEMP_FOLDER, 'DamnHardwareError.dict'))  # 保存下來字典，可以自己設名稱
print(dictionary)
```

```python
corpus = [dictionary.doc2bow(text) for text in corpus] #把corpus從字元形式轉換成矩陣形式
```

```python
# 這些程式碼基本一個不漏、從官網照抄
import logging
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)

from gensim.models import LdaModel

# Set training parameters.
num_topics = 10
chunksize = 2000
passes = 20
iterations = 400
eval_every = None  # Don't evaluate model perplexity, takes too much time.

# Make a index to word dictionary.
temp = dictionary[0]  # This is only to "load" the dictionary.
id2word = dictionary.id2token

model = LdaModel(
    corpus=corpus,
    id2word=id2word,
    chunksize=chunksize,
    alpha='auto',
    eta='auto',
    iterations=iterations,
    num_topics=num_topics,
    passes=passes,
    eval_every=eval_every
)

model.save('OMG.model')  # 把模型儲存到硬碟裡面，名字隨便
```

```python
top_topics = model.top_topics(corpus) #, num_words=20)

# Average topic coherence is the sum of topic coherences of all topics, divided by the number of topics.
avg_topic_coherence = sum([t[1] for t in top_topics]) / num_topics
print('Average topic coherence: %.4f.' % avg_topic_coherence)

from pprint import pprint
pprint(top_topics)
```

```python
pyLDAvis.enable_notebook()
vis = pyLDAvis.gensim.prepare(model, corpus, dictionary)
pyLDAvis.display(vis)
```
